# Calculate ROA for each company-year

## Return on Assets = (Net Income - Preferred Dividends (if any)) / Average Assets

```python
#Generate the average assets for each year
#First define the function of moving average:
f_roll = lambda x: x.rolling(2).mean()
```

```python
def calculate_rolling_mean(x):
    rolling_window = x.rolling(2)
    mean_values = rolling_window.mean()
    return mean_values
```

```python
#Now apply this function to total assets
df['avg_asset']=df.groupby('tic', group_keys=False).at.apply(calculate_rolling_mean)
df
```

```python
#Calculate roa
df['roa'] = (df.ni/df.avg_asset)*100
df
```

# Calculate ROE for each company-year
## Return on Equity = (Net Income - Preferred Dividends (if any)) / Average Stockholders’ Equity

```python
#Now apply this function to total assets
df['avg_ceq']=df.groupby('tic', group_keys=False).ceq.apply(calculate_rolling_mean)
df
```

### Net Income ≠ Money Actually Received by Shareholders

- **Net Income (NI)**: How much profit the company made in a year (total revenue minus total expenses). This is the accounting profit of the company.  
- **Money Actually Received by Shareholders**: Usually only a portion of the net income, because the company may reinvest the profit, pay debts, or retain earnings instead of distributing all as dividends.  

> Therefore, a high ROE indicates strong profitability, but it does **not** mean shareholders immediately receive that money.

---

### 2️⃣ Meaning of ROE

- **Return on Equity (ROE)** tells you how effectively a company uses shareholders' money to generate profit — essentially, the **efficiency of earning**.  
- High ROE → Company is profitable → Investors are more likely to receive future returns, but it does **not** guarantee immediate cash dividends.

---

### 3️⃣ Implications for Investors

- High ROE → Shareholders' money is being used efficiently.  
- Low ROE → Company is slow to generate profit or not using shareholders’ funds effectively.  
- Investors should consider **dividend policy** and **company reinvestment strategy** to understand the actual return.

---

💡 **Summary**: ROE reflects the company's ability to generate profit, not the actual cash received by shareholders. A high ROE only indicates the company is profitable, but shareholders may not get cash immediately.


```python
#Calculate roe
df['roe'] = (df.ni/df.avg_ceq)*100
df
```

# Calculate PMR for each company-year
## Net Profit Margin = (Net Inc – Pref Divs) / Net Sales [Rev]

**Explanation:**

- **Net Income:** The profit a company makes during a period after deducting all costs, expenses, interest, and taxes.  
- **Preferred Dividends:** Payments that must be made to preferred shareholders, subtracted from net income.  
- **Net Sales / Revenue:** The total money the company earns from selling products or services.

**Meaning:**  
- NPM shows how much profit the company earns for every $1 of sales after paying expenses and preferred dividends.  
- For example, NPM = 10% → for every 100 in sales, the company earns 10 dollar of net profit.

**Interpretation of values:**
- **High NPM (≥ 20%)** → Strong profitability, efficient at turning revenue into profit.  
- **Medium NPM (10% - 20%)** → Moderate profitability.  
- **Low NPM (< 10%)** → Weak profitability, costs or expenses are high.

```python
df['pmr'] = df.ni/df.sale
df
```

# Calculate ATR for each company-year
## Asset Turnover = Net Sales [Rev] / Average Total Assets

```python
#Calculate atr
df['atr'] = (df.sale/df.avg_asset)
df
```

**Meaning:**  

- ATR tells you how much sales a company generates for every 1 unit of asset it owns.  
- For example, ATR = 2 → For every 1 dollar of assets, the company generates 2 dollar in sales.

# Calculate Leverage (Equity Multiplier) for each company-year

## Leverage (Equity Multiplier) = Average Total Assets / Average Common Stockholders’ Equity


```python
#Calculate leverage
df['leverage'] = (df.avg_asset/df.avg_ceq)
df
```
