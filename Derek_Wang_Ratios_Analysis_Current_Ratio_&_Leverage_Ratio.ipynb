## Net Working Capital

## Current Assets (act)
Assets that a company can convert to cash, sell, or use within one year.  

**Common examples:**  
- Cash (money in the bank)  
- Accounts receivable (money owed to the company)  
- Inventory (unsold goods)  
- Short-term investments (easily sellable investments)  

**Simple understanding:** Short-term resources the company can use to pay debts or spend.

---

## Current Liabilities (lct)
Debts or obligations a company must pay within one year.  

**Common examples:**  
- Accounts payable (money owed to suppliers)  
- Short-term loans (must be repaid within a year)  
- Wages payable, taxes payable (expenses due soon)  

**Simple understanding:** Short-term debts the company must pay.

```python
df['nwc'] = df['act'] - df['lct']
df.head()
```

## Current Ratio = Current Assets (ACT) / Current Liabilities (LCT)

```python
df['cr'] = df['act']/df['lct']
df.head()
```

- **Current Ratio (CR) > 1**  The company can cover its short-term debts (financially healthy)  
- **Current Ratio (CR) < 1**  The company may not be able to pay its short-term debts (at risk)

## Leverage Ratio (Equity Multiplier) = Average Total Assets (AT) / Average Common Stockholders Equity (CEQ)

```python
#First define the function of moving average:
f_roll = lambda x: x.rolling(2).mean()

#Average Total Assets
df['avg_at'] = df.groupby('tic', group_keys=False)["at"].apply(f_roll)

#Average Common Stockholders Equity
df['avg_ceq'] = df.groupby('tic', group_keys=False)["ceq"].apply(f_roll)
df.head()
```

```python
#Remove year 2010 from the analysis, because it does not have rolling avg data
df = df[df['fyear']>2010]
df.head()
```

```python
#Leverage Ratio
df['leverage_ratio'] = df['avg_at']/df['avg_ceq']
df.head()
```

### Why Leverage Ratio Can Be Negative

The **Leverage Ratio** is calculated as:


Leverage Ratio = Total Assets  Shareholders' Equity

Normally, this ratio is greater than 1 because shareholders' equity is only a part of total assets.  

However, the ratio can become **negative** if **shareholders' equity is negative**, which can happen due to:

- Continuous company losses, making **retained earnings negative**  
- Paying out too much in dividends or **asset devaluation**  
- Accounting adjustments that create negative equity (e.g., negative assets or write-downs)  

When shareholders' equity is negative:

Leverage Ratio = Total Assets  (Negative Equity)  Negative Value

For example, a leverage ratio of **-400** means:

- Shareholders' equity is far smaller than total assets and is negative  
- The company's leverage is extremely high  
- The company is in a very risky financial situation
